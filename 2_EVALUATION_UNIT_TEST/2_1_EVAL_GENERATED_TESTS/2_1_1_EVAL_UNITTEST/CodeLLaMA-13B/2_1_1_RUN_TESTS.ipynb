# Executando os Unit Testes Gerados

```python
outputs = []

for index, row in df.iterrows():
    prompt = row['prompt']
    canonical_solution = row['canonical_solution']
    generated_solution = row['generated_solution']
    
    regex = r"class\s+(\w+)"
    match = re.search(regex, generated_solution)
    test_class_name = match.group(1)
    
    # print(prompt + canonical_solution + generated_solution)
    exec(prompt + canonical_solution + generated_solution)
    print(test_class_name)
    print("--------------------")
    loader = unittest.TestLoader()
    suite = loader.loadTestsFromTestCase(globals()[test_class_name])
    
    runner = unittest.TextTestRunner()
    result = runner.run(suite)

    # print("Total de testes executados:", result.testsRun)
    # print("Testes com falhas:", len(result.failures))
    # print("Testes com erros:", len(result.errors))
    
    correctness = 1 if len(result.failures) == 0 and len(result.errors) == 0 else 0
    
    outputs.append([index, result.testsRun, len(result.failures), len(result.errors), correctness])
```

# Salvando Resultados 

```python
df = pd.DataFrame(outputs, columns=['index', 'tests_run', 'failures', 'errors', 'correctness'])

df.to_csv('eval_test_codellama13b.csv', index=False)
```
