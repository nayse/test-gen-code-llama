```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../../../2_0_PREPROCESSING_INFERENCES/CodeLLaMA-13B/generated_tests_codellama13b_nvidia_api_adjusted.csv',
    sep=',',
    header=0
)
```

```python
def write_file(code, filename):
    with open(filename, 'a') as f:
        f.write(code)
```

```python
import re

for index, row in df.iterrows():

    prompt = row['prompt']
    canonical_solution = row['canonical_solution']
    generated_solution = row['generated_solution']

    regex = r"class\s+(\w+)"
    match = re.search(regex, generated_solution)
    test_class_name = match.group(1)
    
    generated_solution = generated_solution.replace(test_class_name, test_class_name+ "_" + str(index), 1)
    
    write_file(prompt + canonical_solution, "functions.py")
    write_file(generated_solution + "\n\n", "test_functions.py")
    
```
